```python
#WDBC
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import numpy as np
sns.set() #make the plots look pretty
sns.set(rc={"figure.dpi":300, 'savefig.dpi':300})
sns.set_context('notebook')
sns.set_style("ticks")
```

```python
def box_plot(results, title):
    df_nn = pd.DataFrame(results[:, 0])
    df_sn = pd.DataFrame(results[:, 1])
    df_sci = pd.DataFrame(results[:, 2])
    df_sni = pd.DataFrame(results[:, 3])
    df_sfi = pd.DataFrame(results[:, 4])
    accs = pd.concat([df_nn, df_sn, df_sci, df_sni, df_sfi], axis=1)
    accs = accs.set_axis(['NN', 'SN', 'SC', 'SNI', 'SFI'], axis=1, inplace=False)
    sns.boxplot(data=accs, palette= "ch:s=3,rot=-.10")
    plt.ylabel('Solution Cost')
    plt.title(title)
```
